```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\navyashree\Desktop\MIni DataPipeline\salary.csv")
```

```python
print("Original Data")
```

```python
print(df)
```

```python
df = df.drop_duplicates()
print("After Removing duplicates")
print(df)
```

```python
df=df.drop_duplicates(subset=["name", "age", "salary", "department", "city"])
```

```python
print("After Cleaning")
print(df)
```

```python
df.isnull().sum()
```

```python
df["age"] = df["age"].fillna(df["age"].mean())
```

```python
df["salary"] = df["salary"].fillna(df["salary"].median())

```

```python
df["department"]= df["department"].fillna(df["department"].mode()[0])
```

```python
print("Removed missing values at age, salary, department")
print(df)
```

```python
df["age"]=df["age"].astype(int)
```

```python
df["salary"]=df["salary"].astype(int)
```

```python
print(df)
```

```python
df["salary_per_year"] = df["salary"] * 12
```

```python
print(df)
```

```python
df["city"] = df["city"].replace({
    "Bangaluru": "Bangalore",
    "Manglore": "Mangalore",
    "Kopla": "Koppal"
})
```

```python
print(df)
```
